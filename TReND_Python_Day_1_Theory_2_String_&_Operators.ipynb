## **1. Operators**


### What is an operator?

[Operators](https://www.tutorialspoint.com/python/python_basic_operators.htm) are the constructs which can manipulate the value of **operands**.

In programming, an **operand** is a term used to describe any object that is capable of being manipulated. A basic example of an operand is a variable declared. For example, in "4 + 5" the "4" and "5" are the operands and the plus symbol is the **operator**.

**Types of Operator:**
Python language supports the following types of operators.

- Arithmetic Operators *(introduced in Tutorial 1)*
- **Relational or Comparison Operators** *(introduced now)*
- **Assignment Operators** *(introduced now)*
- **Logical Operators** *(introduced now)*
- Bitwise Operators
- Membership Operators
- Identity Operators

When you want to justify one condition while the other condition is not true, then you use Python if else statement.

### **Python Nested if Statement!**
### Basic structure:

   - if **condition_1** :      
     - if **condition_2**:     
           - **action for condition_2**
   - elif **condition_3**:
       - **action for condition_3**
   - else:                    
      - **else action if condition_1 or condition_2 are not met**


We can have a **if...elif...else** statement inside another **if...elif...else** statement. This is called **nesting** in computer programming.

Any number of these statements can be nested inside one another. **Indentation** is the only way to figure out the level of nesting. They can get confusing, so they must be avoided unless necessary.

Set n=95
