## **1. Operators**


### What is an operator?

[Operators](https://www.tutorialspoint.com/python/python_basic_operators.htm) are the constructs which can manipulate the value of **operands**.

In programming, an **operand** is a term used to describe any object that is capable of being manipulated A basic example of an operand would be a variable declared. For example, in "4 + 5" the "4" and "5" are the operands and the plus symbol is the **operator**.

**Types of Operator:**
Python language supports the following types of operators.

- Arithmetic Operators *(introduced in Tutorial 1)*
- **Relational Operators** *(introduced now)*
- **Assignment Operators** *(introduced now)*
- Logical Operators
- Bitwise Operators
- Membership Operators
- Identity Operators

## **1.1 What is a relational operator?**

Relational operators compare the values on either sides of them and decide the relation among them. Written below are examples of relational operators:

- "=="  (equal)
- "!="  (not equal)
- ">"   (greater than)
- "<"   (smaller than)
- ">="  (greater or equal to)
- "<="  (smaller or equal to)

### **The operator "==" (equal)**

If the values of two operands are equal, then the condition becomes true.

```python
number_student_enjoying_swimming=10
number_student_not_enjoying_swimming=10
print("Is the number of students enjoying swimming \
equal to the number of students who do not enjoy swimming?")
print(number_student_enjoying_swimming==number_student_not_enjoying_swimming)
```

### **The operator "!=" (not equal)**

If values of two operands are not equal, then condition becomes true.

```python
number_student_enjoying_swimming=10
number_student_not_enjoying_swimming=20
print("Is the number of students enjoying swimming \
different to the number of students who do not enjoy swimming? ")
print(number_student_enjoying_swimming!=number_student_not_enjoying_swimming)
```

### **The operator ">" (greater than)**

If the value of left operand is greater than the value of right operand, then condition becomes true.

```python
number_student_enjoying_swimming=20
number_student_not_enjoying_swimming=10
print("Is the number of students enjoying swimming \
greater than the number of students who do not enjoy swimming? ")
print(number_student_enjoying_swimming>number_student_not_enjoying_swimming)
```

### **The operator "<" (smaller than)**

If the value of left operand is less than the value of right operand, then condition becomes true.

```python
number_student_enjoying_swimming=10
number_student_not_enjoying_swimming=20
print("Is the number of students enjoying swimming \
smaller than the number of students who do not enjoy swimming? ")
print(number_student_enjoying_swimming<number_student_not_enjoying_swimming)
```

### **The operator ">=" (greater or equal to)**

If the value of left operand is greater than or equal to the value of right operand, then condition becomes true.

```python
number_student_enjoying_swimming=10
number_student_not_enjoying_swimming=10
print("Is the number of students enjoying swimming \
greater or equal to the number of students who do not enjoy swimming?  ")
print(number_student_enjoying_swimming>=number_student_not_enjoying_swimming)
```

### **The operator "<=" (smaller or equal to)**

If the value of left operand is less than or equal to the value of right operand, then condition becomes true.

```python
number_student_enjoying_swimming=10
number_student_not_enjoying_swimming=10
print("Is the number of student enjoying swimming \
 smaller or qual to the number of students who do not enjoy swimming?")
print(number_student_enjoying_swimming<=number_student_not_enjoying_swimming)
```

## **1.2 What is an assignment operator?**

**Assignment operators** are used in Python to assign values to variables, and automatically overwrite the previous value that was assigned to the variable. Written below are some examples of assignment operator:

### **The increment operator "+=" (add and assign)**

It adds right operand to the left operand and assign the result to left operand.

```python
# assume a bus with an initial amount of passengers
number_of_people_on_the_bus=10 
print(number_of_people_on_the_bus)

# at the next bus stop 5 people enter
number_of_people_on_the_bus+=5 
# this increases the number of people on the bus by 5
print(number_of_people_on_the_bus)
```

### **The decrement operator "-=" (substract and assign)**

It subtracts right operand from the left operand and assign the result to left operand.

```python
# the bus continues driving with this passenger load:
print(number_of_people_on_the_bus)

# at the next bus stop 3 people leave the bus
number_of_people_on_the_bus-=3 
# this decreases the number of people on the bus by 3
print(number_of_people_on_the_bus)
```

## **2. Conditional statement (`if`-`else`)**

### What is a conditional statement?

A [conditional statement](https://www.tutorialspoint.com/python/python_loops.htm) in Python allows to perform different computations or actions depending on whether a specific Boolean constraint evaluates to true or false. Conditional statements are handled by `if` statements in Python.

### **What is Python `if` Statement?**

An [`if` conditional statement](https://www.guru99.com/if-loop-python-conditional-structures.html#1) is used for decision-making operations. It contains a body of code which runs only when the condition given in the if statement is true. If the condition is false, then the optional else statement runs which contains some code for the else condition. When you want to justify one condition while the other condition is not true, then you use Python if else statement.

### **What happens when `if` condition is not met?**

```python
# Deborah and Sokhar have both a mango tree, lets see who got more mangoes
mango_deb = 8 #mangoes Deborah has
mango_sok = 4  #mangoes Sokhar has
if (mango_deb < mango_sok):
  st = "Deborah has less mangoes than Sokhar."
print(st)
```

It shows the error because it does not match our "if condition" (i.e. mango_deb < mango_sok), therefore, the variable "st" should  not even exist, so when one ask the computer "show me the value of st", the computer does not actually know what you are talking about. 

### **How to use the `else` condition?**

The `else` condition is usually used when you have to judge one statement on the basis of other. If one condition goes wrong, then there should be another condition that should justify the statement or logic.

```python
# Deborah and Sokhar have both a mango tree, lets see who got more mangoes
mango_deb = 8 #mangoes Deborah has
mango_sok = 4  #mangoes Sokhar has
if (mango_deb < mango_sok):
  st = "Deborah has less mangoes than Sokhar."
else:
  st = "Deborah has more mangoes than Sokhar." 
print(st)
```

Use "else condition", if there is any other outcomes you want to print out in case your "if condition" does not give the expected result.

### **When  the `else` condition does not work ...**

There might be many instances when your `else` condition will not give you the desired result. It will print out the wrong result as there is a mistake in program logic. In most cases, this happens when you have to justify more than two statements or conditions in a program.

An example will better help you to understand this concept.

Here both the variables are same (8,8) and the program output is "mango_deb is greater than mango_sok", which is WRONG. This is because it checks the first condition (if condition in Python), and if it fails, then it prints out the second condition (else condition) as default. In next step, we will see how we can correct this error.

```python
# Deborah and Sokhar have both a mango tree, lets see who got more mangoes
mango_deb = 8 #mangoes Deborah has
mango_sok = 8  #mangoes Sokhar has
if (mango_deb < mango_sok):
  st = "Deborah has less mangoes than Sokhar."
else:
  st = "Deborah has more mangoes than Sokhar." 
print(st)
```

oops!....Now both `mango_deb` and `mango_sok` over here are the same (both Deborah and Sokhar have the same amount of mangoes), still it prints out `Deborah has more mangoes than Sokhar.`.

### **How do we use the `elif` condition?**

To correct the previous error made by "else condition", we can use "elif" statement. By using "elif" condition, you are telling the program to print out the third condition or possibility when the other condition goes wrong or incorrect.

```python
# Deborah and Sokhar have both a mango tree, lets see who got more mangoes
mango_deb = 8 #mangoes Deborah has
mango_sok = 8  #mangoes Sokhar has
if (mango_deb < mango_sok):
  st = "Deborah has less mangoes than Sokhar."
elif (mango_deb == mango_sok):
  st = "Deborah and Sokhar have the same amount of mangoes." 
else:
  st = "Deborah has more mangoes than Sokhar." 
print(st)
```

### **Python Nested if Statement!**

We can have a **if...elif...else** statement inside another **if...elif...else** statement. This is called **nesting** in computer programming.

Any number of these statements can be nested inside one another. **Indentation** is the only way to figure out the level of nesting. They can get confusing, so they must be avoided unless necessary.

Following example demonstrates nested if Statement Python.

```python
total = 100
country = "US"
#country = "AU"
if country == "US":
  if total <= 50:
    print("Shipping Cost is $50")
  elif total <= 100:
    print("Shipping Cost is $25")
  elif total <= 150:
    print("Shipping Costs $5")
  else:
    print("FREE")
if country == "AU":
  if total <= 50:
    print("Shipping Cost is $100")
  else:
    print("FREE")
```
