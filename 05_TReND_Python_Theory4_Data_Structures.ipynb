# **2. What is a list?**

### **2.1 Definition**

- A list is a collection of objects (e.g integers, strings, tuples etc)
- A list can contain duplicate data
- A list can contain different data types
- A list is mutable that is, allows addition, insertion, substitution and deletion
- A list is denoted with an open square bracket ( **[** ) , one or more python objects, 
and a closing square bracket ( **]** )

![List](https://pynative.com/wp-content/uploads/2021/03/python-list.jpg)

### **2.2 Creating a List**

To initialize an empty list, use an open and close bracket [ ] as below, or use the list object declarator, list()

```python
# Declaring an empty list
empty_list1 = []
empty_list2 = list()
```

You can use the `type` function to ask for the type of objects ( data type or data structure ) that is held by a variable

```python
print(type(empty_list1)), print(type(empty_list2))
```

We can also initialize list with data in it, as shown below

```python
# Declaring list with elements
list1 = ['a', 'b', 'c', 'a', 'b', 'c', 'd']                                 # list can contain characters
list2 = ['apple', 'mango', 'banana', 'pineapple', 'avocado', 'orange']      # list can contain strings 
list3 = [2, 3, 4, 5, -2, 4, 10, 50]                                         # list can contain numbers
list4 = ['a', 2, 'boy', 'girl', 'b']                                        # list can contain mixed-type objects
```

To find the total amount of elements in a list, you can compute the length of a list with `len(list)`:

```python
print(len(list2)) # To find the total number of fruits in the list
```

### **2.3 List Indexing and Slicing**

- List indexing and slicing involves picking a particular subset of a list
- List elements are selected using an index/integer or a range
- Lists are zero-indexed i.e. counting starts from 0 **not 1**
- Only integers are allowed as indexes in list

![list indexing](https://cdn.programiz.com/sites/tutorial2program/files/python-list-index.png)

#### **List Indexing**

```python
print(list1)
```

```python
# To select the third element of a list
print(list1[2])                             # index 3 - 1 = 2
```

#### **Negative Indexing**

- In negative indexing/slicing, list is index from the reverse side of the list

- Negative indexing should be used with caution, to avoid errors

```python
# To select the last element in the list
list2[-1]
```

```python
# To select the third to last element in the list
list2[-3]
```

```python
# Select all elements except the last
list2[:-1] # or list2[0:-1] 
```

# **3. What is a dictionary?**

### **3.1 Definition**

 A dictionary is a collection of key-value pairs where each key-value pair maps the key to its associated value.


### **3.2 Creating a Dictionary**

- One way to create a dictionary is to start with the empty dictionary (denoted by "{ }") and add key-value pairs as shown below;

```python
greetings = {} #defining an empty dictionary
        
greetings[1] = 'Welcome' 
greetings[2] = 'To'           # 1,2,3,4,5 are the keys 
greetings[3] = 'the'          # and
greetings[4] = 'Python'       # "Welcome", "To", "The", "Python" , "Workshop" are the values
greetings[5] = 'Workshop'

print(greetings)
```

- Another way to create a dictionary is to provide a bunch of key-value pairs using the same syntax as the previous output.

```python
greetings = {1: 'Welcome', 2: 'To', 3: 'the', 
             4: 'Python', 5: 'Workshop'}
greetings
```

**Note that** it doesnt matter what order we write the pairs. The values in a dictionary are accessed with keys, not with indices.

- You can also create a dictionary using **dict( )**

```python
club_expression = dict(Exciting='Mathematics', Cool='Programming',  
                       Fun='Physics', Awesome='Computer Science')
club_expression
```

### **3.3 Dictionary Methods**

Dictionaries have a number of useful built-in methods. We will discuss some of them here, but more details can be found in the [Python documentation](https://docs.python.org/3/library/stdtypes.html#mapping-types-dict)


```python
club_expression = dict(Exciting='Mathematics', Cool='Programming',  
                       Fun='Physics', Awesome='Computer Science') # defines dictionary

key_list = list(club_expression.keys()) 
print(key_list) # prints all the keys of the dictionary. The keys of a dictionary must be unique

value_list = list(club_expression.values())
print(value_list) # prints all the values of the dictionary

item_list = list(club_expression.items())
print(item_list) # prints all the key-value pairs of the dictionary

```

### **3.4 Indexing a Dictionary**

To index elements in a dictionary, we need to know about the keys in the dictionary.

```python
print(club_expression.keys())
```

```python
club_expression['Exciting']         # should give "Mathematics"
```

```python
club_expression['Awesome']          # should give "Computer Science"
```

## **4.1 [Self Study] What is a tuple?**

### **4.1.1 Definition**
- A tuple, like a list, is a sequence of items of any type. 

- A tuple is represented by a parentheses **( )** instead of square brackets [ ].



### **4.1.2 Creating a Tuple**

Enclose the element/elements in parentheses **( )**, separated by a comma.

```python
persons = ("Warren", "Buffett", 1930, "Rihanna", 1988, "Jack", "Ma", 1964, "Katherine", "Johnson", 1918) #contains mixed datatypes

num_tup = (2, 3, 4, 7) #contains integers only

char_tup = ("Sewade", "Yolande", "Morris", "Ebi") #contains characters only
```

You might also want to create a tuple with a single element

```python
t = (4,)  #notice that there is a comma "," 
```

**Question**: What happens if we omit the comma in the above example?

### **4.1.3 Accessing Elements of a Tuple**


Tuples are indexed and sliced similarly as lists

```python
#Create tuple object and assign it to the variable name
name = ("T", "R", "E", "N", "D", "I", "N", "A", "F", "R", "I", "C", "A") 

print(name[0])  # prints the first element of the tuple

print(name[-1]) # prints the last element of the tuple

print(name[3])   # prints the element at position 3
```

```python
interns_info = ("Rose", 2008, "Daniel", 2018, "Remy", 2001, "Gloria", 2012, "Hubert", 2020)

print(interns_info[:2]) # prints the first 2 elements
print(interns_info[3:]) # prints all elements starting at index 3 onwards
```

```python
# Can you guess what will be printed?
interns_info[:]
```

### **4.1.4 Tuple Features**
 - Tuples are **immutable**. That is, an element of a tuple cannot be assigned a different value once it has been set.
 For example,

```python
new_tup = (2, 3)

new_tup[1] = 4
```

- Tuples can be concatenated. Concatenation means to merge two things together, and this is done using the "+" sign as shown below;

```python
tup1 = (1, 2)
tup2 = (3, 4)
tup3 = tup1 + tup2
print(tup3)

interns_info = interns_info[:4] + ("Cisse", 2010) + interns_info[6:] #using the slice operator defined earlier
print(interns_info)
```

- Python built-in functions can be applied to Tuples.
Tuples support a number of Python built-in functions such as len( ), sum( ), min( ) and max( ), just to cite a few.

Here is a list of [Python Built-in Functions](https://docs.python.org/3/library/functions.html)

```python
date = (2008, 2018, 2001, 2012, 2020)

print(len(date)) # gives the total number of elements in date

print(max(date)) # prints highest date

print(min(date)) # prints the lowest date
```

## **4.2 [Self Study] What is a set?**

### **4.2.1 Definition**

- A set is a collection of unique, unordered objects
- Insertion order into a set is not guaranteed
- Sets are useful data structures because they enable fast lookup

### **4.2.2 Creating a Set**

```python
# Declaring an empty set

empty_set = set()
```

```python
# Initializing a set with elements

print(set(['ad', 'add', 'ad', 'addy']))             # Set can be initialized with list or tuple
```

```python
print(list1) 
print('\nLength of list1 is:', len(list1))
```

```python
set_list1 = set(list1)                              # A set only stores unique elements
print(set_list1)

print('\nLength of set_list1 is:', len(set_list1))
```

```python
print(type(set_list1))
```

The main advantage of set is that it collects unique elements, and **ensure fast lookup of those element through hashing** (advanced)

```python
'y' in set_list1            # check if y is cantained in set. Similar to list
```

```python
print(list3)
```

```python
set_list3 = set(list3)      # Sets can contain any hashable datatype like int, float, str, tuples
print(set_list3)
```

```python
# Alternative method: set_union_1_3 & set_list3
```

```python
print(set_union_1_3)
```

**Convert set to another data structure such as list or tuple**

```python
print(list(set_union_1_3))
```

#### Q1: Collect the age in years of all the people in the list above. If you added names of people not in your breakout room, estmate their age.  Can your represent these ages as a tuple called `ages` and print it?

```python
# @title Double click here for the solution {display-mode: "form"}
# We assume the ages of the 7 persons in your list are given as follows; (12, 14, 15, 17, 19, 21, 12)

ages =  (12, 14, 15, 17, 19, 21, 12,)
print(ages)
```

#### Q2: Print a sorted version of the ages (as a tuple).

```python
# @title Double click here for the solution {display-mode: "form"}

sorted_ages =  sorted(ages)
print(tuple(sorted_ages))
```

#### Q4: Print the mean of the ages.

```python
# @title Double click here for the solution {display-mode: "form"}
mean_age =  sum(ages) / len(ages)
print('Average of participant ages: ', mean_age)
```

#### Q5: What is the minimum and maximum age in the tuple? Print it. 

```python
# @title Double click here for the solution {display-mode: "form"}

minimum_age =  min(ages)

maximum_age = max(ages)

print('Min student age: ', minimum_age)
print('Max student age: ', maximum_age)
```

#### Q6: What is/are the index/indices of the minimum and maximum ages? Print it. How can you deal with it, if the max and minimum age may be found in more than one index?

```python
# @title Double click here for the solution {display-mode: "form"}
# The max and minimum ages may be found in more than one index, 
# so use a list to store all the indexes found

index_min_ages = []
index_max_ages = []

for index, element in enumerate(ages):
    if element == minimum_age:
        index_min_ages.append(index)
    elif element == maximum_age:
        index_max_ages.append(index)

print('Index of min ages: ', index_min_ages)
print('Index of max ages: ', index_max_ages)
```

## Level 3: Trickiest to extra tricky

#### Q1: Assign different ages to the list of participants above by creating a dictionary of the participant as keys and their ages as values. The dictionary should look something like this `{'mark': 12, 'ola': 14, 'sandra': 15, ...}`

```python
# @title Double click here for the solution {display-mode: "form"}
age_dict = {}
for i in range(len(ages)):
    participant = participant_list[i]
    participant_age = ages[i]
    age_dict[participant] = participant_age

print(age_dict) 
```

#### Q2: Who is the youngest participant in your dictionary?

```python
# @title Double click here for the solution {display-mode: "form"}
min_age = min(age_dict.values())
youngest_p = []

for participant, age in age_dict.items():
    if age == min_age:
        youngest_p.append(participant)

print(f"Youngest participant(s) in the class is/are: {youngest_p}")
```

#### Q3: Which participants have the same age?

```python
# @title Double click here for the solution {display-mode: "form"}
age_tracker = {}

same_age_list = []

for participant, age in age_dict.items():
    if age in age_tracker:
        same_age_list.append(participant)
        same_age_list.append(age_tracker[age])
        
        break
    else:
        age_tracker[age] = participant

print("List of participants with the same age: ", same_age_list)
```
